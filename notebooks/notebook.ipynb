# Project Background

Landmine contamination remains a critical humanitarian and developmental challenge in many conflict-affected regions, including Ukraine, the Democratic Republic of Congo, Sudan, Myanmar, Yemen, and Angola. Millions of square kilometres of land remain unsafe due to anti-personnel mines, anti-tank mines, improvised explosive devices (IEDs), and unexploded ordnance (ERW), causing thousands of civilian casualties and restricting agricultural activities, infrastructure development, and the safe return of displaced populations. 

Traditional demining efforts are slow, costly, and resource-intensive, highlighting the urgent need for innovative solutions. Machine learning offers a transformative approach to accelerate landmine detection, risk mapping, and prioritisation of clearance operations. By integrating geospatial data, terrain characteristics, historical conflict information, and soil measurements, ML models can predict high-risk zones, optimise demining resource allocation, and ultimately reduce civilian casualties and economic losses. 

This project seeks to develop robust ML algorithms capable of assisting demining agencies in war-torn and post-conflict regions, thereby supporting humanitarian rehabilitation and safer community resettlement.

# Validation framework

```python

def split_df(df, target, val_ratio=0.2, test_ratio=0.2, random_state=42):
    """
    Splits a dataframe into train, validation, and test sets.
    
    Parameters
    ----------
    df : pd.DataFrame
        The full dataset including the target column.
    target : str
        Name of the target column.
    val_ratio : float, optional
        Fraction for validation set.
    test_ratio : float, optional
        Fraction for test set.
    random_state : int, optional
        Random seed for reproducibility.

    Returns
    -------
    df_train : pd.DataFrame
        Training features.
    df_val : pd.DataFrame
        Validation features.
    df_test : pd.DataFrame
        Test features.
    y_train : np.array
        Training target.
    y_val : np.array
        Validation target.
    y_test : np.array
        Test target.
    """

    # ---  Compute split sizes ---
    n = len(df)
    n_val = int(n * val_ratio)
    n_test = int(n * test_ratio)
    n_train = n - n_val - n_test

    # ---  Shuffle indexes ---
    idx = np.arange(n)
    np.random.seed(random_state)
    np.random.shuffle(idx)

    idx_train = idx[:n_train]
    idx_val = idx[n_train:n_train + n_val]
    idx_test = idx[n_train + n_val:]

    # ---  Split into subsets ---
    df_train = df.iloc[idx_train].reset_index(drop=True)
    df_val = df.iloc[idx_val].reset_index(drop=True)
    df_test = df.iloc[idx_test].reset_index(drop=True)

    # ---  Extract target values ---
    y_train = df_train[target].values
    y_val = df_val[target].values
    y_test = df_test[target].values

    # --- Drop target from X ---
    df_train = df_train.drop(columns=[target])
    df_val = df_val.drop(columns=[target])
    df_test = df_test.drop(columns=[target])

    return df_train, df_val, df_test, y_train, y_val, y_test
```

```python
df_train, df_val, df_test, y_train, y_val, y_test = split_df(df, target='Mine_type')
```

```python
def prepare_X(df, base, new_features_list=None,fill_strategy="median",encode_list=None):
    '''
    Prepares feature matrix X and dictionary representation from dataframe.
    Parameters
    ----------
    df : pd.DataFrame
        Input dataframe.
    base : list
        List of base feature names.
    new_features_list : list, optional
        List of additional feature names to include.
    fill_strategy : str, optional
        Strategy for filling missing values: "median", "mean", or "mode".
    encode_list : list, optional.List of categorical features to encode (not used in current implementation).    
    Returns 
        ------- 
        X : np.array -Feature matrix.
        dicts : list of dicts.List of feature dictionaries for each record.'''
    df = df.copy()

    # --- 1. Construct final list of features ---
    final_features = base.copy()

    if new_features_list:
        for feat in new_features_list:
            if feat not in df.columns:
                raise ValueError(f"Feature '{feat}' not found in DataFrame")

            if feat not in final_features:
                final_features.append(feat)

    # --- 2. Check categorical features ---
    if encode_list:
        for col in encode_list:
            if col not in final_features:
                raise ValueError(
                    f"Column '{col}' must be included in base or new_features_list"
                )

    # --- 3. Handle missing values BEFORE DictVectorizer ---
    for col in final_features:
        if df[col].isna().sum() > 0:

            if fill_strategy == 'median':
                val = df[col].median()

            elif fill_strategy == 'mean':
                val = df[col].mean()

            elif fill_strategy == 'mode':
                val = df[col].mode()[0]

            else:
                raise ValueError("fill_strategy must be median, mean, or mode")

            df[col].fillna(val, inplace=True)

    # --- 4. Select only final columns ---
    df_final = df[final_features]
    X= df_final.values
    dicts = df_final.to_dict(orient="records")
    return X,dicts

```

```python
# Train model
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import LogisticRegression
from catboost import CatBoostClassifier
from sklearn.ensemble import RandomForestClassifier



from sklearn.metrics import accuracy_score


lr = LogisticRegression(C=1.0, max_iter=1000)
catb = CatBoostClassifier(depth=10,learning_rate=0.2,iterations=200,verbose=0,loss_function='MultiClass')
rf = RandomForestClassifier(n_estimators=300,max_depth=8,random_state=42)
models = [lr,catb,rf]

def model_train(y_train,dicts,model_list):

    trained = []    
    dv = DictVectorizer(sparse=False)
    X_train_vec = dv.fit_transform(dicts)
    for model in model_list:
        model.fit(X_train_vec, y_train)
        trained.append(model)
    return dv,trained
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
def svc_train(y_train, dicts,C =1.0):
    
    dv = DictVectorizer(sparse=False)
    scaler = StandardScaler()

    X_train_vec = dv.fit_transform(dicts)
    
    X_train_scaled = scaler.fit_transform(X_train_vec)

    model = SVC(kernel='rbf',C=C, gamma='scale')
    
    model.fit(X_train_scaled, y_train)
    return dv, model
```

```python
def model_evaluate(X,dicts,model_list):
    ''' Evaluates the model on given data.
    Parameters
    ----------
    X : np.array-Feature matrix.
    y_true : np.array
        True target values.
    dicts : list of dicts-List of feature dictionaries for each record.
    model : Trained model-The model to evaluate.
    Returns
    -------------------------------
    y_pred : np.array
        Predicted target values.
    '''
    predicted = {}
    dv = DictVectorizer(sparse=False)
    X_vec = dv.fit_transform(dicts)
    for model in model_list:
        y_pred = model.predict(X_vec)

        predicted[model.__class__.__name__] = y_pred
    return predicted



```

```python

def scores(y_test,pred_dict):
    scores = {} 
    for y_pred_key,y_pred_val in pred_dict.items():
        accuracy = accuracy_score(y_test, y_pred_val)
        scores[y_pred_key] = accuracy
    return scores
```

# Fine tuning 



```python
def hyperparameter_tuning_catboost(X_train, y_train, X_val, y_val, depth_values, learning_rate_values, iteration_values):
    '''Tunes hyperparameters for CatBoostClassifier using grid search.
    Parameters
    ----------
    X_train, y_train,X_val &  y_val  : np.array.
    Training feature matrix,Training target values,Validation feature matrix,& Validation target values,respectively.
    depth_values, learning_rate_values and iteration_values: TYPE list. List of depth values,learning_rate,&iteration_values to try,respectively.
    Returns 
    -------------------------------------
    model : CatBoostClassifier. Trained CatBoost model with best hyperparameters.
    best_params : dict. Best hyperparameters found.
    best_score : float. Best accuracy score achieved.
    ''' 

    best_score = 0
    best_params = {}
    
    for depth in depth_values:
        for learning_rate in learning_rate_values:
            for iterations in iteration_values:
                model = CatBoostClassifier(depth=depth, learning_rate=learning_rate, iterations=iterations, verbose=0, loss_function='MultiClass')
                model.fit(X_train, y_train)
                y_pred = model.predict(X_val)
                accuracy = accuracy_score(y_val, y_pred)
                if accuracy > best_score:
                    best_score = accuracy
                    best_params = {
                        'depth': depth,
                        'learning_rate': learning_rate,
                        'iterations': iterations
                    }
    return model,best_params,best_score
```
